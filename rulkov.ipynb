The Rulkov map is a simple dynamical systems model of neuronal action potentials.
The model is a coupling of a fast and a slow dynamical system in the following form

For some specific parameter choices, this system can reproduce the broad outlines of neuronal
action potentials, viz. bursts of spikes, tonic spiking and periods of silence. I would like you to

```python
import numpy as np
import matplotlib.pyplot as plt

N = 6000

class RulkovMap:

    def __init__(self, N, alpha, sigma, mu):
        self.N, self.alpha, self.mu, self.sigma = N, alpha, mu, sigma
        self.X = np.zeros((1, N))
        self.Y = np.zeros((1, N))
    
    def f(self, n):

        if self.X[:, n] <= 0:
            return self.alpha/(1-self.X[:, n]) + self.Y[:, n]

        elif (self.X[:, n] < (self.alpha + self.Y[:, n])):
            return self.alpha + self.Y[:, n]
            
        else:
            return -1

    def simulate_step(self, n):

        self.X[:, n+1] = self.f(n)
        self.Y[:, n+1] = self.Y[:, n] - self.mu*(self.X[:, n] + 1) + self.mu*self.sigma
```

```python
#bursting spikes
alpha=6
sigma=-0.1
mu=0.001
env = RulkovMap(N, alpha=alpha, sigma=sigma, mu=mu)
for i in range(N-1):
    env.simulate_step(i)

fig, (ax1, ax2) = plt.subplots(2, sharex=True, figsize=(8,4))
fig.suptitle(f'Bursting spikes with Î±={alpha}, Ïƒ={sigma}, Î¼={mu}')
ax1.plot(env.X.T[1500:])
ax1.set(ylabel='x_n')
ax2.plot(env.Y.T[1500:])
ax2.set(xlabel='n', ylabel='y_n')
plt.show()
```
